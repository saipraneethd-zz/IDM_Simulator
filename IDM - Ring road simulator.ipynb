```python
import math
import matplotlib.pyplot as plt
```

```python
class car:
    def __init__(self,_location,_speed):
        self.speed = _speed
        self.location = _location
    def pos_update(self):
        location = self.location + self.speed * 0.1
        return location
    def vel_update(self,leader_position,leader_speed):
        if (leader_position - self.location) > 0:
            s_alpha = (leader_position - self.location) - 5
        else:
            s_alpha = (round((leader_position - self.location),2) % (600)) - 5
        s_star = 2 + (self.speed * 1.5) + (self.speed * (self.speed - leader_speed)/(2*math.sqrt(0.73*1.67)))
#       print(leader_position)
#       print(leader_speed)
        print(s_alpha)
#       print(s_star)
        speed = self.speed + 0.1*0.73* (1-math.pow((self.speed/20),4)-math.pow((s_star/s_alpha),2))
        return speed
```

```python
t = 1000
n = 10
veh = []
veh1_vel = [20]
veh1_time = [0]

## Intializzation
for j in range(0,n):
    veh.append(car(j*60,20))

##Update
for i in range(1,t+1):
    for j in range(0,n):
        print(j)
        if j==0:
            veh1_pos = veh[0].location
            # PLot 
            veh1_vel.append(veh[0].speed)
            veh1_time.append((i) * 0.1)
        if j < n-1:
            veh[j].speed = veh[j].vel_update(veh[j+1].location,veh[j+1].speed)
        else:
            veh[j].speed = veh[j].vel_update(veh1_pos,veh[0].speed)
        veh[j].location = veh[j].pos_update() % (600)
```
